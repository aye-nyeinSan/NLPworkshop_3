```python
import nltk
import pandas as pd
import re

#load the dataset
df = pd.read_csv("Datasets/spam.csv",encoding='latin-1')
print("loaded dataset!")
df
```

```python
# Drop the specified columns
df.drop(columns=["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], inplace=True)

# Display the updated DataFrame
print(df)
```

```python
#remove whitespaces and convert to lowercase 
df['v2'] = df['v2'].apply(lambda x: x.lower())
df['v2'] = df['v2'].apply(lambda x: re.sub(r'\s+', ' ', x))
df['v2']
```

```python
#remove anything that is not English
df['v2'] = df['v2'].apply(lambda x: re.sub(r'[^a-zA-Z]', ' ', x))
df['v2']
```

```python
if 'v1' in df.columns and 'v2' in df.columns:
    df['label'] = df['v1']
    df['text'] = df['v2']
else:
    print("Column 'v1' and 'v2' does not exist in the DataFrame.")
    df.drop(columns=['v1', 'v2'], inplace=True)

df = df[['label', 'text']]
df['length'] = df['text'].apply(lambda x: len(x.split()))
    
df
```

```python
#create new column for removing numbers and special characters
df['text2'] = df['text'].apply(lambda x: re.sub(r'[^a-zA-Z]', ' ', x))
df
```

```python
#labelEncoding the target variable 
#!pip3 install scikit-learn
from sklearn.preprocessing import LabelEncoder

df['label_transform']= LabelEncoder().fit_transform(df['label'])
df
```

```python
#count the count of unique words changing text data  to numerical data
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words=None,token_pattern=r'\b\w+\b')
matrix = vectorizer.fit_transform(df['text2'])

# Convert to array and inspect the result
print(matrix.toarray())  # This should now contain non-zero values
print("Total unique words:", len(vectorizer.get_feature_names_out()))  # Total unique words
```

```python
matrix_array = pd.DataFrame(matrix.toarray(),columns= vectorizer.get_feature_names_out())
matrix_array
```

```python
#Sorting the top 10 words
matrix_array.T.sort_values(0,ascending=False).head(10)
```

```python
#first 5 rows of the transformed dataset
print ("Top 5 rows of the transformed dataset")
matrix_array.head(5)
```

```python
#Bottom 5 rows of the transformed dataset
print ("Bottom 5 rows of the transformed dataset")
matrix_array.tail(5)
```
